```python
import numpy as np
import pandas as pd
import statsmodels.api as sm
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
```

```python
raw_data = pd.read_csv("candy-data.csv")
data = raw_data.drop(["competitorname"],axis=1)
data
```

```python
data.describe()
```

```python
x1 = data.drop(["chocolate"],axis=1)
y = data["chocolate"]
```

```python
from sklearn.model_selection import train_test_split
x_train , x_test , y_train , y_test = train_test_split(x1, y, test_size=0.2,random_state=42)
```

## Training

```python
x_train = sm.add_constant(x_train)
reg_log = sm.Logit(y_train,x_train)
reg_results = reg_log.fit()
reg_results.summary()
```

```python
def confusion_matrix(data,actual_values,model):
        
        # Confusion matrix 
        
        # Parameters
        # ----------
        # data: data frame or array
            # data is a data frame formatted in the same way as your input data (without the actual values)
            # e.g. const, var1, var2, etc. Order is very important!
        # actual_values: data frame or array
            # These are the actual values from the test_data
            # In the case of a logistic regression, it should be a single column with 0s and 1s
            
        # model: a LogitResults object
            # this is the variable where you have the fitted model 
            # e.g. results_log in this course
        # ----------
        
        #Predict the values using the Logit model
        pred_values = model.predict(data)
        # Specify the bins 
        bins=np.array([0,0.5,1])
        # Create a histogram, where if values are between 0 and 0.5 tell will be considered 0
        # if they are between 0.5 and 1, they will be considered 1
        cm = np.histogram2d(actual_values, pred_values, bins=bins)[0]
        # Calculate the accuracy
        accuracy = (cm[0,0]+cm[1,1])/cm.sum()
        # Return the confusion matrix and 
        return cm, accuracy
```

```python
confusion_matrix(x_train,y_train,reg_results)
```

## Testing

```python
x_test = sm.add_constant(x_test)
```

```python
confusion_matrix(x_test,y_test,reg_results)
```

The accuracy for training dataset is 95.55%. (1)
The accuracy for the test dataset is 94.11%.  (2)
We expect that the (2)<(1).

Hence,  we can conclude that our model explains around 94-95% of the variability.
